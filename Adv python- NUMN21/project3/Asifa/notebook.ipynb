### Modeling temperature distribution in a small apartment using the Laplace equation


```python
import numpy as np
from mpi4py import MPI
from scipy.linalg import solve
import sys
import matplotlib as plt
```

### We are simulating how heat spreads in a small apartment made of 3 connected rooms by solving the Laplace equation, and we use MPI to let each room’s temperature be calculated by a different process that communicates with its neighboring rooms until the temperatures balance out.

### Each MPI process is a room — they talk to each other to share heat until the whole apartment reaches steady temperature.
